# Road Accident in india during 2001-2012

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_excel("datafile.xls")
```

```python
x=data.iloc[:,1]
y=data.iloc[:,3]
```

```python
plt.title("male death rate year wise")
plt.bar(x,y,color=["skyblue"],edgecolor='blue')
plt.xlabel("years")
plt.ylabel("number of male death")
plt.show()
```

```python
#More number of male accident at 2010 and 2012
```

```python
plt.title("female death rate year wise")
plt.bar(x,data["Female"],color=["skyblue"],edgecolor='blue')
plt.xlabel("years")
plt.ylabel("number of female death")
plt.show()
```

```python
#More number of female accident at 2006 and 2008
```

```python
df=data.groupby(["States"]).agg({"Total":sum}).head(20)
df.plot(kind="barh",fontsize=10)
plt.title("state wise total death rate")
plt.xlabel("number of deaths")
plt.ylabel("States")
plt.show()
```

```python
#more number of accidents are occures at andhrapredesh
#the second more number of accidents are occures at karnataka
```

```python
df=data.groupby(["States"]).agg({"Male":sum}).head(20)
df.plot(kind="barh",fontsize=10)
plt.title("state wise total Male death rate")
plt.xlabel("number of Male deaths")
plt.ylabel("States")
plt.show()
```

```python
#the more number of male accident in andhrapredesh
```

```python
df=data.groupby(["States"]).agg({"Female":sum}).head(20)
df.plot(kind="barh",fontsize=10)
plt.title("state wise total Female death rate")
plt.xlabel("number of Female deaths")
plt.ylabel("States")
plt.show()
```

```python
#the more number of female accident in andhrapredesh
```

```python
cause=data.groupby(["CAUSE"]).agg({"Total":sum})
cause.plot(kind="barh",fontsize=10)
plt.title("causes based total death rate")
plt.xlabel("number of deaths")
plt.ylabel("different causes")
plt.show()
```

```python
#the truck/lorry(private) is make maximum accidents
#the second maximum accident is make by two wheelers
```

```python
cause=data.groupby(["CAUSE"]).agg({"Male":sum})
cause.plot(kind="barh",fontsize=10)
plt.title("causes based total male death rate")
plt.xlabel("number of male deaths")
plt.ylabel("different causes")
plt.show()
```

```python
#the truck/lorry(private) is make maximum  male accidents
#the second maximum male accident is make by two wheelers
```

```python
cause=data.groupby(["CAUSE"]).agg({"Female":sum})
cause.plot(kind="barh",fontsize=10)
plt.title("causes based total female death rate")
plt.xlabel("number of female deaths")
plt.ylabel("different causes")
plt.show()
```

```python
#the truck/lorry(private) is make maximum female accidents
#the second maximum female accident is make by two wheelers
```

```python
new_dataframe = data[["Year","Male","Female"]]
var = new_dataframe.groupby(['Year']).sum()
var.plot(kind='bar',stacked=False,grid=False)
plt.xlabel('Deaths', fontsize = 8)
plt.ylabel('Number of deaths gender wise', fontsize=8)
plt.title("Stacked Death Genderwise over the year")
plt.show()
```

```python
#every year number of female death rate is lower than the male death rate
#80% male death rate is equal to 20% female death rate
```

```python
new_dataframe = data[["CAUSE","Male","Female"]]
var = new_dataframe.groupby(['CAUSE']).sum()
var.plot(kind='barh',stacked=False,grid=False)
plt.xlabel('Deaths', fontsize = 8)
plt.ylabel('Number of deaths gender wise', fontsize=8)
plt.title("Stacked Death Genderwise over the year")
plt.show()
```

```python
fig = plt.figure(figsize=(40,40))
data_frame_andhra = data[data["States"] == "TAMIL NADU"]
df_Andhdra =data_frame_andhra[["CAUSE","Male","Female"]]
df= df_Andhdra.groupby(['CAUSE']).sum()
df.plot(kind="barh", fontsize = 10)
plt.xlabel('number of deaths', fontsize = 10)
plt.title("Deaths Cause in TAMILNADU")
plt.show ()
```

```python
#the maximum number of male accident by truck/lorry in tamilnadu
#the second maximum number of male accident by two wheeler
```
