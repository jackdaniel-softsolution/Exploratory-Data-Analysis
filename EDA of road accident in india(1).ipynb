# Road Accident in india during 2001-2012

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_excel("datafile.xls")
```

```python
plt.rcParams["figure.figsize"]=(15,5)
new_dataframe = data[["Year","Male"]]
var = new_dataframe.groupby(['Year']).sum()
var.plot(kind='bar',stacked=False,grid=False)
plt.title("The Male death rate in a year wise",color="green",fontsize="25")
plt.xticks(rotation='vertical')
plt.ylabel("Number of Male death",fontsize="20")
plt.xlabel("Years",fontsize="20")

plt.show()
```

```python
#More number of male accident occurs at 2010,2011 and 2012
```

```python
plt.rcParams["figure.figsize"]=(15,5)
new_dataframe = data[["Year","Female"]]
var = new_dataframe.groupby(['Year']).sum()
var.plot(kind='bar',stacked=False,grid=False)
plt.title("The Female death rate in a year wise",color="green",fontsize="25")
plt.xticks(rotation='vertical')
plt.ylabel("Number of Female death",fontsize="20")
plt.xlabel("Years",fontsize="20")
plt.show()
```

```python
#More number of female accident occurs at 2010,2011 and 2012
```

```python
plt.rcParams["figure.figsize"]=(13,8)
df=data.groupby(["States"]).agg({"Total":sum}).head(20)
df.plot(kind='barh',stacked=False,grid=False)
plt.title("States Wise total death rate",color="green",fontsize="25")
plt.ylabel("States",fontsize="20")
plt.xlabel("Number of Deaths",fontsize="20")
plt.show()
```

```python
#Highest  number of accident occurs at Andhra Pradesh
```

```python
plt.rcParams["figure.figsize"]=(13,8)
df=data.groupby(["States"]).agg({"Male":sum}).head(20)
df.plot(kind='barh',stacked=False,grid=False)
plt.title("States Wise total Male death rate",color="green",fontsize="25")
plt.ylabel("States",fontsize="20")
plt.xlabel("Number of Deaths",fontsize="20")
plt.show()
```

```python
#Highest  number of Male accident occurs at Andhra Pradesh
```

```python
plt.rcParams["figure.figsize"]=(13,8)
df=data.groupby(["States"]).agg({"Female":sum}).head(20)
df.plot(kind='barh',stacked=False,grid=False)
plt.title("States Wise total Female death rate",color="green",fontsize="25")
plt.ylabel("States",fontsize="20")
plt.xlabel("Number of Deaths",fontsize="20")
plt.show()
```

```python
#Highest  number of Female accident occurs at Andhra Pradesh
```

```python
plt.rcParams["figure.figsize"]=(13,8)
cause=data.groupby(["CAUSE"]).agg({"Total":sum})
cause.plot(kind="barh",fontsize=10)
plt.title("Causes based death rate",color="green",fontsize="25")
plt.ylabel("Causes",fontsize="20")
plt.xlabel("Number of Deaths",fontsize="20")
plt.show()
```

```python
#The maximum accidents are occurred by the truck/lorry(private)
#The second maximum accidents is occures by two-wheelers
```

```python
plt.rcParams["figure.figsize"]=(13,8)
cause=data.groupby(["CAUSE"]).agg({"Male":sum})
cause.plot(kind="barh",fontsize=10)
plt.title("Causes based death rate",color="green",fontsize="25")
plt.ylabel("Causes",fontsize="20")
plt.xlabel("Number of Deaths",fontsize="20")
plt.show()
```

```python
plt.rcParams["figure.figsize"]=(13,8)
new_dataframe = data[["Year","Male","Female"]]
var = new_dataframe.groupby(['Year']).sum()
var.plot(kind='bar',stacked=False,grid=False)
plt.title("Stacked Death Genderwise over the year",color="green",fontsize="25")
plt.ylabel("Number of Death rate",fontsize="20")
plt.xlabel("Year",fontsize="20")
plt.show()

```

```python
#Every year number of female death rate is lower than the male death rate
#80% Male death rate is equal to 20% female death rate
```

```python
plt.rcParams["figure.figsize"]=(13,8)
new_dataframe = data[["CAUSE","Male","Female"]]
var = new_dataframe.groupby(['CAUSE']).sum()
var.plot(kind='barh',stacked=False,grid=False)
plt.title("Causes based death rate",color="green",fontsize="25")
plt.xlabel("Number of Death rate",fontsize="20")
plt.ylabel("Causes",fontsize="20")
plt.show()
```

```python
plt.rcParams["figure.figsize"]=(13,8)
data_frame_andhra = data[data["States"] == "TAMIL NADU"]
df_Andhdra =data_frame_andhra[["CAUSE","Male","Female"]]
df= df_Andhdra.groupby(['CAUSE']).sum()
df.plot(kind="barh",stacked=False,grid=False)
plt.title("Tamilnadu causes based death rate",color="green",fontsize="25")
plt.xlabel("Number of Death rate",fontsize="20")
plt.ylabel("Causes",fontsize="20")
plt.show()

```
